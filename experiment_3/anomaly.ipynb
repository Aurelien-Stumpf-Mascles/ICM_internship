```python
import numpy as np
import matplotlib.pyplot as plt
import skimage
```

```python
img = np.zeros((256, 256, 256))
radius_anomaly = 10

random_pixel = np.random.randint(radius_anomaly, 256-radius_anomaly, (3))
```

```python
nb_points = 10
img = np.zeros((100, 100, 100))
radius_anomaly = 20
radius = 50//nb_points
list_centers = []
centre_initial = np.random.randint(radius_anomaly, 100-radius_anomaly, (3))
list_centers.append(centre_initial)

for i in range(nb_points):
    random_idx = np.random.randint(len(list_centers))
    center = list_centers[random_idx]
    random_phi = np.random.rand() * np.pi
    random_theta = np.random.rand() * 2 * np.pi
    new_center = center + (2*radius*np.array([np.sin(random_phi) * np.cos(random_theta), np.sin(random_phi) * np.sin(random_theta), np.cos(random_phi)])).astype(int)
    list_centers.append(new_center)
```

```python
for center in list_centers:
    img[center[0], center[1], center[2]] = 1
```

```python
import skimage.morphology


img = skimage.morphology.dilation(img,footprint=skimage.morphology.ball(radius_anomaly))

img = skimage.morphology.convex_hull_image(img)
```

```python
# display 3D image
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.voxels(img)
plt.show()
```
